```r
library(nnet)
library(party)
library(caret)
library(party)
library(randomForest)
library(e1071)
```

# 인공신경망

```r
ANN <- function(tr,x,y){
    model.nnet <- nnet(등급 ~ ., data = tr, size=3)
    # 검증하기
    pred <- predict(model.nnet, x, type = "class")

    # 예측률 
    return(mean(pred ==y))
}
```

# 의사결정나무

```r
DMT <- function(tr,x,y){
    model.ctree <- ctree(등급 ~ ., data=tr)
    
    # 검증하기
    pred <- predict(model.ctree, x)
    
    # 예측률 
    return(mean(pred ==y))
}
```

# RandomForest

```r
RF <- function(tr,x,y){
    model.rf <- randomForest(등급 ~ ., data = tr , ntree=100)

    # 검증하기
    pred <- predict(model.rf, x)    
    # 예측률 
    return(mean(pred ==y))
}
```

# SVM

```r
SVM <- function(tr,x,y){
    model.svm = svm(등급 ~ ., data=tr)
    # 검증하기
    pred <- predict(model.svm, x)
    # 예측률 
    return(mean(pred ==y))
}
```

# 로지스틱회귀분석

```r
LRA <- function(tr,x,y){
    model <- multinom(등급 ~ .,data=tr)
    # 검증하기
    pred <- predict(model, x)
    # 예측률
    return(mean(pred ==y))
}
```

```r
반복수행 <- function(n){
    irr <- iris

    samp <- createDataPartition(bodyPerformance$등급, p=0.7, list=F)

    irrS <- data.frame(scale(bodyPerformance[,-5]))
    irrS$등급 <- bodyPerformance$등급

    tr <- irrS[samp,]
    test <- irrS[-samp,]

    x <- subset(test, select = -등급)
    y <- test$등급

    
    annS <- 0
    dmtS <- 0
    rfS <- 0
    svmS <- 0
    lraS <- 0
    
    for(i in 1:n){
        annS = annS + ANN(tr,x,y)
        dmtS = dmtS + DMT(tr,x,y)
        rfS = rfS + RF(tr,x,y)
        svmS = svmS + SVM(tr,x,y)
        lraS = lraS + LRA(tr,x,y)       
    }
    
    예측률 <- c(annS/n, dmtS/n, rfS/n, svmS/n, lraS/n)
    
    rtn = data.frame(예측률)
    rownames(rtn) = c('인공신경망','의사결정나무','랜덤포레스트','SVM','로지스틱회귀분석')
    
    return(rtn)

}
```

```r
반복수행(10)
```
