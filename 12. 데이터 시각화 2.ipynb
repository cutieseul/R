### 한 화면에 차트여러개 그려서 비교하기



```r
price <- runif(10, 1, 100)
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해 보기

options(warn=-1)
par(mfrow = c(2,2))
plot(price, type='l')
barplot(price, type='o')
plot(price, type='h')
pie(price)
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해 보기

par(mfrow = c(2,2))
plot(price, type="o", pch=5)
plot(price, type="o", pch=15)
plot(price, type="o", pch=20, col='blue')
plot(price, type="o", pch=c(rep(15,8),20,20)
                    ,col = 'orange'
                    ,cex = c(rep(1,8),3,1))   # 크기주는거
```

### 중첩 자료 시각화

```r
x <- c(1,2,3,4,2,4)
y <- rep(2,6)
```

```r
plot(x,y)
# 데이터 중복
```

```r
# Method 1) 중복데이터 크기로 나타내기 : 겹치거나 가림
xy.df <- as.data.frame(table(x,y))
xy.df
plot(x,y,
    pch = 15, # 네모
    cex = 2 * xy.df$Freq)
# 크기땜에 옆에꺼 가릴수 있어서 잘 안씀
```

```r
# Method 2) 중복데이터 나타내기 

x <- c(1,1,1,2,2,2,2,2,2,3,3,4,5,6)
y <- c(1,1,4,2,3,2,2,2,3,2,1,1,1,1)

z <- data.frame(x,y)
sunflowerplot(z)

# iris$Sepal.Length와 iris$Sepal.Width의 중복데이터 
a <- iris$Sepal.Length
b <- iris$Sepal.Width
공주 <- data.frame(a,b)
sunflowerplot(공주)
```

```r
election <- read.csv('../Data/election_2017.csv')
par(family = 'AppleGothic', bg = '#fafafa')
options(scipen=100)

정 <- t(arrange(election, desc(pop)))[,1:5]
정
name <- paste(정[2, 1:5], 정[3, 1:5])

문 <- as.numeric(정[5, 1:5])
홍 <- as.numeric(정[6, 1:5])
안 <- as.numeric(정[7, 1:5])

합 <- 문+홍+안
율 <- rbind(round(문/합*100,2), round(홍/합*100,2), round(안/합*100,2))
율
데이터 <- as.matrix(정[5:7, 1:5])
rownames(데이터) <- c('문재인', '홍준표', '안철수')

bp <- barplot(
  데이터,
  names = name,
  cex.names = 0.6,
  ylim = c(0,500000),
  main = '투표수 높은 지역의 후보별 투표수',
  legend.text = T,
  col = c('skyblue', 'red', 'lightgreen')
)
for(i in 1:5){
  text(0.7 + 1.2*(i-1), 합[i]+12000, 합[i], cex=1)
  text(0.7 + 1.2*(i-1), 문[i]*0.5, paste(율[1,i],"%"), cex=1)
    text(0.7 + 1.2*(i-1), 문[i]+홍[i]*0.5, paste(율[2,i],"%"), cex=1)
    text(0.7 + 1.2*(i-1), 문[i]+홍[i]+안[i]*0.5, paste(율[3,i],"%"), cex=1)
}
```

# 수학 함수의 시각화
y=x^2

```r
f <- function(x){
    y = x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
x <- seq(-5,5,0.1)
# y=x^2 그래프
plot(x,f(x),
    type="l")
```

y = e^(-x) -4x^2 = x +3|x| -3

```r
f <- function(x){
    y = exp(-x) -4 * x^2 + x + 3 * abs(x) -3
}
```

```r
plot(x, f(x),
    type='l')
```
