```r
seoul <- read.csv("../R/newSeoul2019.csv")
head(seoul)
```

```r
# row names 정리하기
rownames(seoul) <- c()
head(seoul)
```

```r
# factor 정리하기
seoul$자치구명 <- factor(seoul$자치구명)
str(seoul)
```

```r
str(seoul)
```

```r
# 발생건수와 부상자수 상관계수 
cor(seoul$발생건수, seoul$부상자수)
```

```r
# 발생건수와 부상자수의 산포도와 회귀선 그리기
par(family = "AppleGothic")
options(warn=-1)
plot(seoul$부상자수 ~seoul$발생건수,
    main = "발생건수와 부상자수의 산포도")
r <- lm(seoul$부상자수 ~seoul$발생건수 )
abline(r, col ='red') 
options(warn=-1)
```

```r
# 예측력
summary(r)
```

- p-value: < 2.2e-16
- Adjusted R-squared:  0.9621
- Residual standard error: 15.23
