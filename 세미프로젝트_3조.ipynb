```r
상권 <- read.csv("../Data/2018-2021관광특구별상권.csv")
a <- subset(상권,년도==c("2020"))
b <- subset(상권,년도==c("2021"))
코로나 <- rbind(a,b)
코로나
```

```r
covid <- read.csv("../Data/분기별서울시확진자수.csv", header = T)[,-1]

```

```r

# 매출금액 비율 구하기
a1 <- c()

for(j in seq(1,36,5)){
  for(i in j:(j+4)){

    a1 <- append(a1, 코로나$매출금액[i]/sum(코로나$매출금액[j:(j+4)]))
  }
}

b1 <- c()
for(i in 1:8){
  b1 <- append(b1, rep(covid[i,3],5))
}
 round(a1*b1,2)
```

```r
상권$매출금액대비확진자수 <- c(rep(0,40), round(a1*b1,2))
library(dplyr)
```

```r
코로나 <- 상권[41:80,]
# arrange(코로나,년도,관광특구)
par(family = "AppleGothic")
options(scipen = 100)
options(warn=-1)
name = c("2020년1분기","2020년2분기","2020년3분기","2020년4분기",
        "2021년1분기","2021년2분기","2021년3분기","2021년4분기")
               


이태원 <- subset(코로나,관광특구 =="이태원")[,4]
명동 <- subset(코로나,관광특구 =="명동 남대문 북창동 다동 무교동")[,4]
동대문패션타운 <- subset(코로나,관광특구 =="동대문패션타운")[,4]
종로청계 <- subset(코로나,관광특구 =="종로 청계")[,4]
잠실 <- subset(코로나,관광특구 =="잠실")[,4]



plot(이태원/100000000,
     type="o",
     ann = F, # x,y의 label 제거
     axes = F, # x,y좌표 frame 제거
     ylim=c(0,70),
     col = "forestgreen",
     pch = 23
       )
title(main="2020-2021 분기별, 관광특구별 매출현황")
axis(1,at=1:8,lab=name,cex.axis=0.5)
axis(2, cex.axis=0.7)


# y축 title(label) 지정
title(xlab = "년도,분기별", col.lab = "blue")
title(ylab = "매출금액 [단위:억(원)]", col.lab = "blue")
lines(명동/100000000,
     type = "o",
     col = "orange",
     lty = 1,
      pch = 22
      
     )
lines(동대문패션타운/100000000,
     type = "o",
     col = "red",
     lty = 1,
      pch = 21
     )
lines(종로청계/100000000,
     type = "o",
     col = "purple",
     lty = 1,
      pch = 25
     )
lines(잠실/100000000,
     type = "o",
     col = "skyblue",
     lty = 1,
      pch = 24
     )
legend(6, 70, # legend 박스 위치 좌표
   c("동대문","명동","이태원","잠실","종로청계"),
    col = c("red","orange","forestgreen","skyblue","purple"),
    pch = c(21,22,23,24,25),
    lty = c(1,2,2),
    cex = 0.8 # 크기부여 기본값이 1임
) 


이태원확진 <- subset(코로나,관광특구 =="이태원")$매출금액대비확진자수
명동확진 <- subset(코로나,관광특구 =="명동 남대문 북창동 다동 무교동")$매출금액대비확진자수
동대문패션타운확진 <- subset(코로나,관광특구 =="동대문패션타운")$매출금액대비확진자수
종로청계확진 <- subset(코로나,관광특구 =="종로 청계")$매출금액대비확진자수
잠실확진 <- subset(코로나,관광특구 =="잠실")$매출금액대비확진자수

plot(이태원확진,
     type="o",
     ann = F, # x,y의 label 제거
     axes = F, # x,y좌표 frame 제거
     ylim=c(0,35000),
     col = "forestgreen",
     pch = 23
       )
title(main="2020-2021 분기별, 관광특구별 확진자수")
axis(1,at=1:8,lab=name,cex.axis=0.5)
axis(2, cex.axis=0.7)



title(xlab = "년도,분기별", col.lab = "blue")
title(ylab = "확진자수", col.lab = "blue")

lines(명동확진,
     type = "o",
     col = "orange",
     lty = 1,
      pch = 22
     )
lines(동대문패션타운확진,
     type = "o",
     col = "red",
     lty = 1,
      pch = 21
     )
lines(종로청계확진,
     type = "o",
     col = "purple",
     lty = 1,
      pch = 25
     )
lines(잠실확진,
     type = "o",
     col = "skyblue",
     lty = 1,
      pch = 24
     )

legend(1, 35000, # legend 박스 위치 좌표
   c("동대문","명동","이태원","잠실","종로청계"),
    col = c("red","orange","forestgreen","skyblue","purple"),
    pch = c(21,22,23,24,25),
    lty = c(1,2,2),
    cex = 0.8 # 크기부여 기본값이 1임
) 
```

```r

# 매출금액 비율 구하기
a1 <- c()

for(j in seq(1,36,5)){
  for(i in j:(j+4)){

    a1 <- append(a1, 코로나$점포수[i]/sum(코로나$점포수[j:(j+4)]))
  }
}

b1 <- c()
for(i in 1:8){
  b1 <- append(b1, rep(covid[i,3],5))
}
 round(a1*b1,2)
상권$점포수대비확진자수 <- c(rep(0,40), round(a1*b1,2))
```
