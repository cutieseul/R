```r
x <- runif(100)
y <- runif(100)

# plot의 조건
    # pch : y > 0.5이면 1, 아니면 18
    # col : y > 0.5이면 "red", 아니면 "blue"
plot(x,y,
    xlim = c(min(x),max(x)),
    ylim = c(min(y),max(y)),
    pch = ifelse(y>0.5,1,18),
    col = ifelse(y>0.5,"red","blue")
)
```

## 선그래프

```r
# 3종류 dataset의 선 그래프

abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
par(family = 'AppleGothic')
plot(abc,
     type = "o",
     col = "red",
     ylim = c(0, 400),
     axes = F, # x,y좌표 frame 제거
     ann = F # x,y의 label 제거
     )
# x축과 x축의 라벨을 설정
axis(1, at =1:5, lab=c('서울','대전','대구','광주','원주')) # axis()는 축만드는 함수
# y축과 y축의 라벨을 설정
axis(2, ylim = c(0,400))

# main title 지정 따로빼서 쓰면 컬러값을 줄 수가 있어서 그럼
title(main="과일 판매량", col.main = "red")

# x축 title(label) 지정
title(xlab = "지역", col.lab = "blue")

# y축 title(label) 지정
title(ylab = "판매량", col.lab = "blue")

# 선그리기 함수로 def 그리기
lines(def,
     type = "o",
     pch = 22, # 사각형으로 보인다.
     col = "green",
     lty = 2
     )

# 선그리기 함수로 ghi 그리기
lines(ghi,
      type= 'o', 
      pch = 22,
      col = 'blue',
      lty = 2
     )

# 범례 legend를 써줘야됨
legend(4, 400, #x,y좌표
       c("야구장","축구장","해변가"),
    col = c("red","green","blue"),
    pch = c(21,22,22),
    lty = c(1,2,2),
    cex = 0.8 # 크기부여 기본값이 1임
    )
```

# 막대그래프

```r
# 막대그래프는 dataframe을 모름, matrix만 알아듣는다!
x <- 1:6 #x는 그냥 벡터이기 때문에 names를 부여해줘야된다.
barplot(x,
        main = "1~6",
        xlab = "Sales",
        ylab = "Amounts",
        names = c("A","B","C","D","E","F"),
        col = c(rep("red",3),rep("blue",3))
        
       )
```

```r
# 복잡한 형태의 막대 그래프
# 3행 2열의 matrix 생성(dataframe일 시 as.matrix로 바꿔줘야된다.)
x <- matrix(1:6, 3, 2)
x
barplot(x,
        ylim = c(0, 16)
       )
```

```r
# 누적 막대 그래프 분리후 열별 label 붙이기

par(family = "AppleGothic")
#rownames로 만들어줘야 범례를 쓸 수 있기 때문에 만들기
rownames(x) <- c('1월','2월','3월')

barplot(x,
        ylim= c(0,8),
        beside = T, # 누적된 막대의 해체
        names = c("korea","America"),
#         names = c("k1","k2","k3","A1","A2","A3",)
        main = "수출량",
        xlab = "국가",
        ylab = "빈도수",
        col = c("red","green","blue"),
        legend.text = T
       )
```

# 학력별 점수 차이

```r
data <- read.csv("../Data/education.csv", header = F)
head(data)
tail(data)
str(data)
summary(data)
```

```r
# V1 : 남,여
# V2 : 점수
# V3 : 국,영,수
# V4 : 초,중,고,대,원

# 문자,숫자 둘다 가능한 factor로 바꿔주기 V2는 숫자기에 X
data$V1 <- factor(data$V1, levels = 1:2, labels = c('남','여'))
data$V3 <- factor(data$V3, levels = 1:3, labels = c('국어','영어','수학'))
data$V4 <- factor(data$V4, levels = 1:5, labels = c('초','중','고','대','원'))
```

```r
head(data)
str(data)
```

```r
# 컬럼 이름 변경(성별, 점수, 과목, 학력)

colnames(data) = c('성별','점수','과목','학력')
head(data)
```

```r
# 막대 그래프로 점수 빈도수를 표현
    # 빈도수 : table()
table(data$점수)
```

# 파이차트

```r
sales <- c(210,110,400,550,600,130)
pie(sales)
```

```r
par(family = 'AppleGothic')
sales <- c(210,110,400,550,600,130)
week <- c('월','화','수','목','금','토')
# 퍼센트 구하기 
ratio <- sales / sum(sales) * 100
label <- paste(week, "\n", ratio, "%")
```

```r
par(family = 'AppleGothic')
sales <- c(210,110,400,550,600,130)
week <- c('월','화','수','목','금','토')
# 퍼센트 구하기 
ratio <- sales / sum(sales) * 100
label <- paste(week, "\n", ratio, "%")

par(family = 'AppleGothic')
pie3D(sales,
       main = "영업실적 파이차트",
       labels = label,
       explode = 0.05,
       labelcex = 0.6
       )

# pie3D: 기본값이 무지개색임
```
